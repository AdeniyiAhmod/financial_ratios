```python
def calculate_current_ratio(df):
    # Extract rows for Total Current Assets and Total Current Liabilities
    current_assets_row = df[df.index == 'Total Current Assets']
    current_liabilities_row = df[df.index == 'Total Current Liabilities']

    # Extract relevant columns for the calculation (years 2016 to 2022)
    years = [2016, 2017, 2018, 2019, 2020, 2021, 2022]
    current_assets_values = current_assets_row[years].values.flatten()
    current_liabilities_values = current_liabilities_row[years].values.flatten()

    # Calculate current ratio for each year
    current_ratio_values = current_assets_values / current_liabilities_values

    # Create a new DataFrame to store the results
    current_ratio_df = pd.DataFrame({
        'Year': years,
        'Current Ratio': current_ratio_values
    })

    return current_ratio_df

# Example usage
current_ratio_df = calculate_current_ratio(industry_avg_pivot)
print(current_ratio_df)
```

```python
def calculate_ratios(df, ratios_to_calculate):
    ratios_dict = {}

    # Calculate ratios for each specified pair
    for ratio_name, (numerator_row, denominator_row) in ratios_to_calculate.items():
        numerator_row_data = df[df.index == numerator_row].values.flatten()
        denominator_row_data = df[df.index == denominator_row].values.flatten()

        # Check if either numerator or denominator rows are not present
        if not numerator_row_data.any() or not denominator_row_data.any():
            print(f"Error: {numerator_row} or {denominator_row} not found.")
            ratio_values = [np.nan] * len(df.columns)  # Return NaN values for the ratio
        else:
            ratio_values = numerator_row_data / denominator_row_data

        ratios_dict[ratio_name] = ratio_values

    # Create a DataFrame from the dictionary
    ratios_df = pd.DataFrame(ratios_dict)
    ratios_df.insert(0, 'Year', df.columns)

    return ratios_df

# Example usage
ratios_to_calculate = {
    'Current Ratio': ('Total Current Assets', 'Total Current Liabilities'),
    'Quick Ratio': ('Total Current Assets - Inventory', 'Total Current Liabilities')
}

ratios_df = calculate_ratios(industry_avg_pivot, ratios_to_calculate)
print(ratios_df)
```
